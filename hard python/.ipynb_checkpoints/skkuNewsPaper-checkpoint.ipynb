```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
url = 'https://www.skku.edu/skku/campus/skk_comm/news.do'
url
```

```python
response = requests.get(url)
response
```

```python
html = response.text # 텍스트를 그냥 긁어온다.
html 
```

```python
soup = BeautifulSoup(html, 'html.parser') # html처럼 정갈하게 긁어온다.
soup
```

```python
title = soup.find_all('div', class_ = 'news_listCont')
title
```

```python
titles = []
for i in title:
    global titles
    titles.append(re.sub(r"[\n\t\s]*", "", i.find("a").get_text()))
    
    # a 태그를 문자열로 치환
    
titles
```

```python
for i in range(0, len(titles)):
    print(i+1, "번 뉴스:", titles[i])
```
