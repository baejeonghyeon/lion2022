```python
date = pd.date_range('20180101', '20181231')
date[-2]
```

```python
Time = []
for time in date:
    Time.append(date.strftime("%Y%m%d"))

Time[0]
```

```python
time_list = Time[0].tolist()
time_list
```

```python
url_list = []

for i in time_list:
    url = "http://www.kobis.or.kr/kobisopenapi/webservice/rest/boxoffice/searchDailyBoxOfficeList.json?key=" + key + "&targetDt="
    date_url = url + i
    url_list.append(date_url)
    
url_list # 1년간 url을 모아놓은 list
```

```python
boxoffice_database = pd.DataFrame()
type(boxoffice_database)
```

```python
# 데이터프레임 구축 전에 json으로 구조가 어떻게 되어 있는지 살펴보자.

test_request = requests.get(url_list[0])
test_json = test_request.json()
test_json

#따라서 ['boxOfficeResult']['dailyBoxOfficeList']로 접근해야 함을 알 수 있음.
```

```python
for i in url_list:
    request = requests.get(i)
    url_json = request.json()
    
    have_to_be_appended = url_json['boxOfficeResult']['dailyBoxOfficeList']
    
    daily_dic = pd.DataFrame.from_dict(have_to_be_appended)
    boxoffice_database = boxoffice_database.append(daily_dic)
#     if i == 'http://www.kobis.or.kr/kobisopenapi/webservice/rest/boxoffice/searchDailyBoxOfficeList.json?key=a8e597633acca03091ad5fba598b58a7&targetDt=20180106':
#         break : 테스트용


boxoffice_database
```

```python
import openpyxl

boxoffice_database.to_csv("boxoffice.csv", encoding="UTF-8", index = False)
boxoffice_database.to_excel("boxoffice.xlsx", index = False)
```
